```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# reading the Excel file
file_path = 'Data Analysis - Caselet#2.xlsx'  # Replace with the actual path to your file
df = pd.read_excel(file_path, sheet_name='DifferentFormat')

# customizing appearance
plt.figure(figsize=(14, 8))
sns.set(style="whitegrid")  # Set the style of the seaborn plot

# creating boxplots using seaborn
sns.boxplot(x='Team Member', y='TPT', data=df,  notch=True, showfliers=False)
plt.title('Boxplots of TPT for Each Team Member')
plt.xlabel('Team Member')
plt.ylabel('TPT Values')

# rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# showing the plot
plt.show()
```
